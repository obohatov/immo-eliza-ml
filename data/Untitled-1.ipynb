```python
df['surface_land_sqm'].fillna(0, inplace=True)
df['terrace_sqm'].fillna(0, inplace=True)
df['garden_sqm'].fillna(0, inplace=True)
df.to_csv('dataset1.csv', index=False)
```

```python
filename = "dataset1.csv"
df = pd.read_csv(filename)
nan_counts = df.isna().sum()

print(nan_counts)
```

```python
df['latitude'] = df['latitude'].interpolate('linear')
df['longitude'] = df['longitude'].interpolate('linear')

df['construction_year'] = df['construction_year'].fillna(df['construction_year'].median())

df['total_area_sqm'] = df['total_area_sqm'].fillna(df['total_area_sqm'].mean())
df['primary_energy_consumption_sqm'] = df['primary_energy_consumption_sqm'].fillna(df['primary_energy_consumption_sqm'].mean())
df['cadastral_income'] = df['cadastral_income'].fillna(df['cadastral_income'].mean())

df['nbr_frontages'] = df['nbr_frontages'].fillna(df['nbr_frontages'].mode())
```

```python
df['nbr_frontages'] = df['nbr_frontages'].fillna(df['nbr_frontages'].median())
nan_counts = df.isna().sum()

print(nan_counts)
```

```python
df.to_csv('dataset2.csv', index=False)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
df['construction_year'] = imputer.fit_transform(df[['construction_year']])

imputer = SimpleImputer(strategy='mean')
df['total_area_sqm'] = imputer.fit_transform(df[['total_area_sqm']])
```

```python
filename = "dataset2.csv"
df = pd.read_csv(filename)

categorical_features = ['fl_furnished', 'fl_open_fire', 'fl_terrace', 'fl_garden', 'fl_swimming_pool', 'fl_floodzone', 'fl_double_glazing']

df = pd.get_dummies(df, columns=categorical_features, drop_first=True)

df.head()
```
